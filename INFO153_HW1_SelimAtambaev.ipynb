## 1. Stack ADT and DS Implementation

The goal of this assignment is to check whether an HTML document has balanced opening and closing tags.  
This problem is very similar to the parentheses matching problem studied in class, where each opening symbol must have a corresponding closing symbol.

HTML tags follow a nested structure, meaning the last opened tag must be the first one to be closed.  
For example, if we open `<html>`, then `<body>`, then `<h1>`,  

we must close them in the reverse order:

`</h1>`, `</body>`, and finally `</html>`.

Because of this behavior, I chose to use a Stack data structure.  
A Stack follows the LIFO (Last In, First Out) principle, which perfectly matches how HTML tags are processed.

In my solution, I implemented a Stack class in Python.  
This class allows me to store opening tags as they are encountered and retrieve them in reverse order when processing closing tags.

Using a Stack makes it easy to keep track of the correct order and ensures that each closing tag matches the most recent opening tag.

```python
class Stack:
    """ sa4252
    Simple stack class for storing HTML tags
    """

    def __init__(self):
        # initialize empty stack
        self.items = []

    def push(self, item):
        # add item
        self.items.append(item)

    def pop(self):
        # remove last item
        if not self.is_empty():
            return self.items.pop()
        return None

    def is_empty(self):
        # check if empty
        return len(self.items) == 0
```

## 2. html_checker Function

This function checks whether the HTML tags are balanced.
First, the HTML string is split into lines. Then each line is examined to determine whether it contains a tag.
If an opening tag is found, it is pushed onto the stack. If a closing tag is found, it is compared with the most recent opening tag stored in the stack.
If the tags do not match, the function immediately returns False. If all tags are processed and matched correctly, the function returns True.

```python
def html_checker(html_string):
    """ sa4252
    checks if html tags are balanced
    """

    stack = Stack()
    html_lines = html_string.split('\n')

    for line in html_lines:
        line = line.strip()

        # skip text
        if not line.startswith("<") or not line.endswith(">"):
            continue

        # closing tag
        if line.startswith("</"):
            tag = line.replace("</", "").replace(">", "")

            if stack.is_empty():
                return False

            top_tag = stack.pop()

            if top_tag != tag:
                return False

        else:
            # opening tag
            tag = line.replace("<", "").replace(">", "")
            stack.push(tag)

    if not stack.is_empty():
        return False

    return True
```
