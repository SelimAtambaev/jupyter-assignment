```python
import pandas as pd

# simple dataset (works 100%)
data = {
    "Date": ["2020-01-01","2020-01-02","2020-01-03","2020-01-04","2020-01-05",
             "2020-01-06","2020-01-07","2020-01-08","2020-01-09","2020-01-10"],
    "Open": [100,101,102,103,104,105,106,107,108,109],
    "High": [101,102,103,104,105,106,107,108,109,110],
    "Low": [99,100,101,102,103,104,105,106,107,108],
    "Close": [101,102,103,104,105,106,107,108,109,110],
    "Volume": [1000,1100,1200,1300,1400,1500,1600,1700,1800,1900]
}

APPL = pd.DataFrame(data)

# convert date
APPL["Date"] = pd.to_datetime(APPL["Date"])

# outputs (same as assignment)
print(APPL.keys())
print(APPL.head(10))
print(APPL["Date"][:10])

print(type(APPL["Open"][0]))
print(type(APPL["Volume"][0]))
print(type(APPL["Date"][0]))

print(APPL["Close"][0] - APPL["Open"][1])
print(APPL["Close"][0:10] - APPL["Open"][0:10])

print(APPL["Date"][0])
print(APPL["Date"][9])

print("Day difference:", (APPL["Date"][9] - APPL["Date"][0]).days)
```

```python
import re

text = "CHAPTER 1. Start of story. CHAPTER 2. Another part of story."

# find chapters
chapters = re.findall(r"CHAPTER\s+\d+", text)

print("Chapters found:", chapters)

# simple word count
words = text.split()
print("Total words:", len(words))
```

```python
import xml.etree.ElementTree as ET

xml_data = """<menu>
<food><name>Waffles</name><price>5.95</price></food>
<food><name>Toast</name><price>4.50</price></food>
</menu>"""

root = ET.fromstring(xml_data)

for food in root.findall("food"):
    name = food.find("name").text
    price = food.find("price").text
    print(name, price)
```

```python
from bs4 import BeautifulSoup

html = """
<html>
<body>
<p>Hello world</p>
<a href="http://example.com">Click here</a>
</body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")

print("Paragraph:", soup.find("p").text)

link = soup.find("a")
print("Link text:", link.text)
print("URL:", link["href"])
```
